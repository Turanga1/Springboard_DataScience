# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df.loc[(hospital_read_df['Number of Discharges'] != 'Not Available')]
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
df = clean_hospital_read_df.sort_values('Number of Discharges')
print(df.info())
```

```python
def conditions(x):
    if x > 800:
        return "large"
    elif x < 350:
        return "small"
    else:
        return "medium"

func = np.vectorize(conditions)
hosp_size = func(df['Number of Discharges'])
df['size'] = hosp_size
```

```python
sns.lmplot(x='Number of Discharges', y='Excess Readmission Ratio', data=df[81:-3], hue='size', scatter_kws={"s": 2}, size=8)
plt.ylim(0.5, 2)
plt.title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)
plt.show()
```

```python
# separate red and green areas:

small_hosp = df['Excess Readmission Ratio'].loc[(df['size'] == 'small')][81:-3]
large_hosp = df['Excess Readmission Ratio'].loc[(df['size'] == 'large')][81:-3]

t, p = weightstats.ztest(small_hosp, large_hosp)
print(t, p)
```

The calculated p value is significantly less than the specified significance $\alpha$ = .01. Therefore we can state that there is a statistically significant difference in excess readmission ratios between large and small hospitals. 

Although the difference is statistically significant, there may be no practical difference. Note that the initial evaluation used three different definitions for "small" hospitals (100, 300 and 350 discharges). Depending on the size of the hospital, the number of "problematic" hospitals can be as low as zero:

```python
# how many "problematic" hospitals are there?

all_hospitals = df['size'].loc[(df['Number of Discharges'])][81:-3].count()
small100 = df['size'].loc[(df['Number of Discharges'] < 100) & (df['Excess Readmission Ratio'] >= 1.15)][81:-3].count()
small300 = df['size'].loc[(df['Number of Discharges'] < 300) & (df['Excess Readmission Ratio'] >= 1.15)][81:-3].count()
small350 = df['size'].loc[(df['Number of Discharges'] < 350) & (df['Excess Readmission Ratio'] >= 1.15)][81:-3].count()

size_df = pd.DataFrame({'hospital size by number of discharges': ['100', '300', '350'], \
                        '% of all hospitals': [small100/all_hospitals*100, small300/all_hospitals*100, small350/all_hospitals*100]},\
                       index=[0, 1, 2])

sns.barplot(x='hospital size by number of discharges', y='% of all hospitals', data=size_df, order=['350', '300', '100'])
plt.title('% hospitals by hospital size for readmission ratio >= 1.15', fontsize=12)
plt.show()
```
